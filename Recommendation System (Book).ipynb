# User Based Collaborative Filtering (UBCF)

```python
user_sim = 1 - pairwise_distances( data.values,metric='cosine') 
user_sim 
```

```python
user_sim_df = pd.DataFrame(user_sim)
user_sim_df 
#converting array to dataframe
```

```python
np.fill_diagonal(user_sim, 0) 
```

```python
user_sim_df 
```

# Item Based Collaborative Filtering (IBCF)

```python
ChildBks_user=data['ChildBks']
ChildBks_user
```

```python
similar_to_ChildBks = data.corrwith(ChildBks_user) 
similar_to_ChildBks
```

```python
ChildBks_corr = pd.DataFrame(similar_to_ChildBks, columns=['Correlation'])
ChildBks_corr
```

```python
ChildBks_corr[ChildBks_corr['Correlation'] > 0].sort_values(by='Correlation', ascending=False)
```

```python
ChildBks_corr[ChildBks_corr['Correlation'] > 0.2].sort_values(by='Correlation', ascending=False) 
```

```python
GeogBks_user=data['GeogBks']
GeogBks_user
```
